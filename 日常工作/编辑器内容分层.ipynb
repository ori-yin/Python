```python
# 定义降维函数
def 降维(x):
    if x == 0:
        return '0次'
    elif x >= 1 and x <= 10:
        return '1-10次'
    elif x >= 11 and x <= 50:
        return '10-50次'
    elif x >= 51 and x <= 100:
        return '51-100次'
    elif x >= 101 and x <= 2000:
        return '101-2000次'
    elif x >= 2000:
        return '2000以上'
    else:
        return '未知'
```

```python
# 将时间差转换为文本型函数
df['编辑区间'] = df['编辑人数'].apply(降维)
```

```python
# 统计不同字符串类型的出现次数
counts = df['编辑区间'].value_counts()

# 绘制条形图
counts.plot(kind='bar')
plt.xlabel('编辑区间')
plt.ylabel('Count')
plt.show()
```

```python
# 将时间差转换为文本型函数
df['下载区间'] = df['下载人数'].apply(降维)
```

```python
# 统计不同字符串类型的出现次数
counts = df['下载区间'].value_counts()

# 绘制条形图
counts.plot(kind='bar')
plt.xlabel('下载区间')
plt.ylabel('Count')
plt.show()
```

# 算价值分

```python
df['下载率'] = np.where(df['编辑人数'] != 0, df['下载人数'] / df['编辑人数'], 0)
```

```python
df['下载人数归一'] = (df['下载人数'] - df['下载人数'].min()) / (df['下载人数'].max() - df['下载人数'].min())
df['编辑人数归一'] = (df['编辑人数'] - df['编辑人数'].min()) / (df['编辑人数'].max() - df['编辑人数'].min())
df['下载率归一'] = (df['下载率'] - df['下载率'].min()) / (df['下载率'].max() - df['下载率'].min())
```

```python
df['价值'] = df['下载人数归一'] * 0.6 + df['编辑人数归一'] * 0.3 + df['下载率归一'] * 0.2
```

```python
df_sort = df.sort_values(by='价值', ascending=False)
df_sort
```

# 抽样

## 3抽1

```python
# 等距抽样
df_sampled = df_sort.iloc[::3]
```

## 2.8抽1

```python
sample_size = int(len(df_sort) / 2.8)

# 生成等距索引
indices = [int(i * 2.8) for i in range(sample_size)]
df_sampled2 = df_sort.iloc[indices]
```

```python
df_sampled2   # 企业素材
```

## 获得差集

```python
merged_df = pd.merge(df, df_sampled2, how='left', indicator=True)
# 选择在表A中但不在表B中的数据
result_df = merged_df.loc[merged_df['_merge'] == 'left_only']
result_df
```

```python
result_df
```

```python
# 判断表A中的数据是否在表B中，并将结果存储在布尔类型的Series中
is_in_b = df['ID'].isin(df_sampled2['ID'])

# 选择在表A中存在但不在表B中的数据
result_df2 = df[~is_in_b]
result_df2
```
