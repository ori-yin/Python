```python
txt = list(df['标题'])
```

```python
b = ','
```

```python
c = b.join(txt)
```

```python
cut = jieba.lcut(c)
```

```python
dict1 = {}
```

```python
for i in cut:
    if len(i) == 1:
        continue
    else:
        dict1[i] = dict1.get(i,0) + 1
```

```python
x = sorted(dict1.items(),key = lambda x:x[1],reverse=True)
```

```python
x2 = pd.DataFrame(data = x,columns=['关键词','次数'],index = None)
```

```python
x2.to_csv('D:/奥里/摄图网词频.csv')
```
